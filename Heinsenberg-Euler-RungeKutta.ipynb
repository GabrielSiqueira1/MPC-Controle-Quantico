Para os processos derivativos é de suma importância, atestar a eficiência numérica e para tal compara-se os métodos de Euler e de Runge-Kutta.

## Euler - Método simples

```python
# Procedimentos iniciais

I = np.zeros((2,2))
H = np.zeros((2,2), dtype=np.complex_)

A = (1/2*np.pi)
B = (3/2*np.pi)

I = [[1, 0], [0, 1]]
H = [[A, 0], [0, B]]
```

```python
# Euler: (\psi(t+delta(t))-\psi(t))/delta t

Psi_0 = [[complex(0.80,0)],[complex(0.60,0)]] # Ponto de partida
Psi_k = np.zeros((2,1000), dtype=np.complex_) # Armazenamento dos pontos
sup = np.zeros((2,1), dtype=np.complex_) 

t = 0

while int(round(t*100, 1)) < 1000: # teste de 10 segundos
    
    deltaT = 0.01
    sup = np.matmul((I - np.dot(deltaT,np.dot(complex(0,1),H))),Psi_0)

    Psi_k[0][int(round(t*100, 1))] = sup[0][0]
    Psi_k[1][int(round(t*100, 1))] = sup[1][0]
    
    Psi_0 = sup
    
    t = t + 0.01
```

```python
ts = np.linspace(0, 10, 1000) #Simulação de 10 segundos

Real = [ele.real for ele in Psi_k[0]] 
Imag = [ele.imag for ele in Psi_k[0]] 

plt.scatter(ts, Imag) 
plt.ylabel('Imaginário') 
plt.xlabel('Tempo') 
plt.show()
```

```python
plt.scatter(ts, Real) 
plt.ylabel('Real') 
plt.xlabel('Tempo') 
plt.show()
```

## Runge-Kutta - Método aprimorado de quarta-ordem

```python
P = np.zeros((2,1000))
S = P.astype(complex)

def dpsidt(t,psi, H):
    return -1*np.matmul(np.dot(complex(0,1),H),psi)

def rungeKutta(t0, psi0, t, h, H):
    
    n = (int)((t-t0)/h)
    
    for i in range(1, n + 1):
        
        S[0][i-1] = psi0[0][0]
        S[1][i-1] = psi0[1][0]
        
        k1 = dpsidt(t0, psi0, H)
        k2 = dpsidt(t0 + 0.5 * h, psi0 + np.dot((0.5*h), k1), H)
        k3 = dpsidt(t0 + 0.5 * h, psi0 + np.dot((0.5*h), k2), H)
        k4 = dpsidt(t0 + h, psi0 + np.dot(h, k3), H)
 
        psi0 = psi0 + (h / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4)
    
        t0 = t0 + h
        
    return S
```

```python
tt = 10
t = 0

h = 0.01

psi0 = [[complex(0.80,0)],[complex(0.60,0)]]

E = np.zeros((2,2))
I = np.zeros((2,2))
R = np.zeros((2,2))
H = E.astype(complex)

A = (1/2*np.pi)
B = (3/2*np.pi)

I = [[1, 0], [0, 1]]
R = [[1, 0], [0, 1]]
H = [[A, 0], [0, B]]

Z = rungeKutta(t, psi0, tt, h, H)
```
