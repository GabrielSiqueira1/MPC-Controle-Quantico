Nesse arquivo, iremos analisar como a equação de Schrodineger-Heisenberg representa os resultados ao longo do tempo sem a necessidade do uso do espaço, apenas dos estados energéticos. Veja que com o avanço do tempo, ocorre uma oscilação que sempre estará limitada ao coeficiente que multiplica a função de onda.

```python
# Bilbiotecas para auxílio na programação matemática
import math, sys 
import numpy as np
import sympy as sp

from scipy import sparse # Produção das diagonais das matrizes
from scipy.sparse import diags 

# Plotagem 2D e 3D
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm


from os import path # Suficiente para manipulação de arquivos
    
# Para solução exata
from scipy.special import hermite
from math import factorial

%matplotlib inline
count = 0

# Para otimização dos sistemas
from scipy import optimize
```

```python
Psi_t = np.zeros((2,2))
Psi_tc = Psi_t.astype(complex)

t = np.linspace(0, 10, 1000) #Simulação de 10 segundos

A = np.exp(-(complex(0,1))*np.pi*t/2)
B = np.exp(-3*(complex(0,1))*np.pi*t/2)

Psi_tc = [[0.80*A], [0.60*B]]

x = [ele.real for ele in Psi_tc[0]] 
y = [ele.imag for ele in Psi_tc[0]] 
plt.scatter(t, y) 
plt.ylabel('Imaginário') 
plt.xlabel('Tempo') 
plt.savefig('imag0.png', format='png')
plt.show()
```
